```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
WCDM17 = pd.read_csv('./data/WCDM_1701.csv', names = ['mx', 'fx'])
WCDM23 = pd.read_csv('./data/WCDM_2309.csv', names = ['mx', 'fx'])
```

```python
mxs = np.logspace(-5, 2, 100)
fxs = [mx/105 for mx in mxs]
```

```python
fig, ax = plt.subplots(figsize=(5,5))

ax.plot(WCDM17.mx, WCDM17.fx, 'r^')
ax.plot(WCDM23.mx, WCDM23.fx, 'r')
ax.fill_between(WCDM23.mx, WCDM23.fx, 1, alpha=0.1, color='r')
ax.plot(mxs, fxs, 'k--')

ax.set_xscale('log')
ax.set_yscale('log')

ax.set_ylim(1e-3,1)
ax.set_xlim(1e-1, 1e3)

ax.set_ylabel(r'$f_{3/2}$', fontsize=18)
ax.set_xlabel(r'$m_{3/2} \mathrm{~} (\mathrm{eV})$', fontsize=18)
```
